## 05. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer).

```python
# Split the data into groups based on the “user_id” column.
# Apply the transform() function on the “order_number” column to generate the maximum orders for each user.
# Create a new column, “max_order,” into which you’ll place the results of your aggregation.

ords_prods_merge['max_order']=ords_prods_merge.groupby('user_id')['number_of_orders'].transform(np.max)
```

```python
# View updated data frame
ords_prods_merge.head(20)
```

```python
# Assign loyalty flag
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check loyalty flags
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# check the newly added column 'loyalty_flag'

ords_prods_merge.head()
```

```python
# Check Loyalty flag in dataframe
ords_prods_merge[['user_id', 'loyalty_flag', 'number_of_orders']].head(20)
```

```python
# the min max and mean of each loyalty flag
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

## 06. find the average prices of items people are buying to determine different types of spenders

```python
# Split the data into groups based on the "user_id" column.
# Apply the transform() function on the "prices" column to generate the maximum orders for each user.
# Create a new column, 'average_spend'into which you’ll place the results of your aggregation.

ords_prods_merge['average_spend']=ords_prods_merge.groupby('user_id')['prices'].transform(np.mean)
```

```python
# Checking the new column
ords_prods_merge.head(20)
```

```python
# Creating flag for low and high spender

ords_prods_merge.loc[ords_prods_merge['average_spend'] < 10, 'spender_flag'] = 'Low_Spender'
ords_prods_merge.loc[ords_prods_merge['average_spend'] >= 10, 'spender_flag']  = 'High_spender'
```

```python
# Check spender flags
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'average_spend', 'spender_flag', 'prices']].head(30)
```

## 7.  Flag for frequent users

```python
# Split the data into groups based on the "user_id" column.
# Apply the transform() function on the "days_since_prior_order" column to generate the maximum orders for each user.
# Create a new column, “ Customer_frequency ” into which you’ll place the results of your aggregation.
```

```python
ords_prods_merge['Customer_frequency']=ords_prods_merge.groupby('user_id')['days_since_prior_order'].transform(np.median)
```

```python
# Cheaking for new column
ords_prods_merge.head()
```

```python
# Assign loyalty flag
ords_prods_merge.loc[ords_prods_merge['Customer_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['Customer_frequency'] <= 20) & (ords_prods_merge['Customer_frequency'] > 10), 'frequency_flag'] = 'Regular customer'                     
ords_prods_merge.loc[ords_prods_merge['Customer_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
                     
```

```python
# Check frequency flags
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
#new value is np.nan. This is a function from the NumPy library; it’s used to assign missing values. 
ords_prods_merge.loc[ords_prods_merge['prices'] >100, 'prices'] = np.nan
```

```python
#Checking for columns
ords_prods_merge[['user_id', 'Customer_frequency', 'frequency_flag']].head(20)
```
