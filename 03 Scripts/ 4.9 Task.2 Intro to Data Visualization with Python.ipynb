### Q3


```python
# Create histogram plot with all bins shown
hist_order_hour_of_day = df_ords_prods_fin['time_of_day'].plot.hist(bins=24, range=(0, 24), color='skyblue', edgecolor='black', alpha=0.7)

# Set axis labels and title
hist_order_hour_of_day.set_ylabel('Orders')
hist_order_hour_of_day.set_xlabel('Hour of Day')
hist_order_hour_of_day.set_title('Distribution of Orders by Hour of Day')

# Set x-axis ticks for all bins
plt.xticks(range(24))

# Show the plot
plt.show()

# export histogram

hist_order_hour_of_day.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'histogram_order_hour_4.9_task.png'))
```

This histogram illustrates the frequency of orders for each hour, revealing that orders are most frequent between 10am and 4pm, while they are least frequent between midnight and 6am.

```python
# Update chart with better formatting:

plt.figure(figsize=(10, 6))  # Adjust figure size
bar_loyalty = df_ords_prods_fin['loyalty_flag'].value_counts().sort_index().plot(kind='bar', color=['red', 'green', 'blue'])
plt.title('Distribution of Loyalty Flags', fontsize=18)  # Increase title font size
plt.xlabel('Loyalty Flag', fontsize=14)
plt.ylabel('Number of Customers', fontsize=14)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)  # Add grid lines for better readability
plt.tight_layout()
plt.show()
```
